```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Create a sample dataset with student performance data
data = {
    'Student_ID': np.arange(1, 101),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Math_Score': np.random.randint(50, 100, 100),
    'Reading_Score': np.random.randint(50, 100, 100),
    'Writing_Score': np.random.randint(50, 100, 100)
}

# Create a DataFrame
student_performance = pd.DataFrame(data)

# Save the dataset as a CSV file
student_performance.to_csv('students_performance.csv', index=False)
print("Dataset 'students_performance.csv' created successfully.")
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('students_performance.csv')

# Explore data
print("First few rows:\n", df.head())
```

```python
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Descriptive statistics
print("Mean:\n", df[['Math_Score', 'Reading_Score', 'Writing_Score']].mean())
print("\nMedian:\n", df[['Math_Score', 'Reading_Score', 'Writing_Score']].median())
print("\nMode:\n", df[['Math_Score', 'Reading_Score', 'Writing_Score']].mode().iloc[0])
print("\nRange:\n", df[['Math_Score', 'Reading_Score', 'Writing_Score']].max() - df[['Math_Score', 'Reading_Score', 'Writing_Score']].min())
```

```python
print("\nStandard Deviation:\n", df[['Math_Score', 'Reading_Score', 'Writing_Score']].std())
```

```python
# Code for Exercise: Calculate Variance
print("\nVariance:\n", df[['Math_Score', 'Writing_Score']].var())
```

```python
import matplotlib.pyplot as plt

# Histograms for data distribution
df[['Math_Score', 'Reading_Score', 'Writing_Score']].hist(bins=10, figsize=(10, 5))
plt.suptitle("Score Distributions")
plt.show()
```

```python
# Box plot for identifying outliers
plt.figure(figsize=(8, 6))
df[['Math_Score', 'Reading_Score', 'Writing_Score']].boxplot()
plt.title("Score Distribution Box Plot")
plt.show()
```
